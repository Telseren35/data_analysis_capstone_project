```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Spotify_Youtube.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Url_spotify','Uri','Url_youtube','Title','Channel','Description'],axis=1, inplace=True)
df
```

```python
df['Artist'].nunique()
```

```python
duplicated_rows = df[df.duplicated(subset=['Artist','Track'])]
print(duplicated_rows)
```

```python
df.drop_duplicates(subset=['Artist','Track'],inplace=True)
```

```python
df
```

```python
df["Duration_ms"] = df["Duration_ms"] / 60000
df
```

```python
album_type_counts = df['Album_type'].value_counts()
album_type_counts
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(x=album_type_counts.index, y=album_type_counts.values, palette='viridis')
plt.title('Albüm Tipine Göre Sayım')
plt.xlabel('Albüm Tipi')
plt.ylabel('Sayım')
plt.xticks(rotation=45)
plt.show()
```

```python
top_artists =df.groupby('Artist')['Stream'].sum().reset_index()
top_10_artists = top_artists.sort_values(by='Stream', ascending=False).head(10)
top_10_artists
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Stream', y='Artist', data=top_10_artists, palette='viridis')
plt.title('En Fazla Stream\'e Sahip Sanatçılar')
plt.xlabel('Toplam Stream Sayısı')
plt.ylabel('Sanatçı')
plt.show()
```

```python
song_durations = df.groupby('Duration_ms').size().reset_index(name='song_count')
```

```python
song_durations
```

```python

plt.figure(figsize=(7,6))
plt.hist(df['Duration_ms'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Şarkı Sürelerinin Dağılımı (Dakika)')
plt.xlabel('Şarkı Süresi (Dakika)')
plt.ylabel('Şarkı Sayısı')
plt.xlim(0, 20)
plt.grid(True)
plt.show()
```

```python
top_likes_youtube = df.groupby('Artist')['Likes'].sum().reset_index()
top_likes_youtube = top_likes_youtube.sort_values(by='Likes', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Likes', y='Artist', data=top_likes_youtube, palette='viridis')
plt.title('En Fazla beğeniye sahip sanatçılar')
plt.xlabel('Toplam Beğeni Sayısı')
plt.ylabel('Sanatçı')
plt.show()
```

```python
top_views_youtube = df.groupby('Artist')['Views'].sum().reset_index()
top_views_youtube = top_views_youtube.sort_values(by='Views', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Views', y='Artist', data=top_views_youtube, palette='viridis')
plt.title('En Fazla Görüntülenmeye sahip sanatçılar')
plt.xlabel('Toplam görüntülenme Sayısı')
plt.ylabel('Sanatçı')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Views", y="Likes")

# Eksen ve başlık ayarları
plt.xlabel('Görüntülenme Sayısı')
plt.ylabel('Beğeni Sayısı')
plt.title('YouTube Videolarında Görüntülenme ve Beğeni Sayısının Karşılaştırması')
plt.grid(True)
plt.show()
```
